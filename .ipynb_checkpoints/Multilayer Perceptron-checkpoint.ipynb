# MLP (Multilayer Perceptron)

```python
# libraries
%matplotlib inline
from __future__ import print_function
from tensorflow.examples.tutorials.mnist import input_data
from sklearn import metrics
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import timeit
sess = tf.InteractiveSession()
```

```python
# load MNIST data
mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)

# placeholder for input data
x = tf.placeholder(tf.float32, [None, 784])
# placeholder to input the correct answers
y_ = tf.placeholder(tf.float32, [None, 10])
```

### Multilayer Perceptron
- activation function: sigmoid
- neurons: 625
- softmax regression
- learning evaluation: cross entropy

```python
def init_weights(shape):
    return tf.Variable(tf.random_normal(shape, stddev=0.01))

def model(X, w_h, w_o):
    h = tf.nn.sigmoid(tf.matmul(X, w_h))
    return tf.matmul(h, w_o)

w_h = init_weights([784, 625])
w_o = init_weights([625, 10])

py_x = model(x, w_h, w_o)
cross_entropy = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(py_x, y_)) # compute costs
```

### Gradient Descent Optimizer:

```python
train_step = tf.train.GradientDescentOptimizer(0.5).minimize(cross_entropy)
# get accuracy
correct_prediction = tf.equal(tf.argmax(py_x,1), tf.argmax(y_,1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

### Training:
- training steps: 15000
- step: 100 samples

```python
# initialize variables
sess.run(tf.initialize_all_variables())

# train
accs = []
start_time = timeit.default_timer()
for i in range(15000):
  batch_xs, batch_ys = mnist.train.next_batch(100)
  acc, _ = sess.run([accuracy, train_step], feed_dict={x:batch_xs, y_: batch_ys})
  accs.append(acc)
elapsed = timeit.default_timer() - start_time
print("Time: %.2f sec" % elapsed)
prediction = tf.argmax(py_x, 1).eval(feed_dict={x: mnist.test.images})
real_labels = tf.argmax(mnist.test.labels, 1).eval()

# results
print("\nAccuracy on test data: %.4f" % accuracy.eval(feed_dict={x: mnist.test.images, y_: mnist.test.labels}))
print()
plt.figure(figsize=(16, 6))
plt.title("Learning rate")
plt.plot(accs)
plt.show()
```

### Plot weights

```python
weights = sess.run(tf.transpose(tf.matmul(w_h, w_o)), feed_dict={x: mnist.train.images, y_: mnist.train.labels})
plt.figure(figsize=(16, 6))
for i in range(len(weights)):
    plt.subplot(2, 5, i+1)
    plt.axis('off')
    plt.imshow(weights[i].reshape(28, 28), cmap=plt.cm.plasma, interpolation='nearest')
    plt.title('Digit: %i' % i)
```

#### Confusion matrix:

```python
print(metrics.confusion_matrix(prediction, real_labels, range(10)))
```

#### Classification report:
- precision: truePositives / (truePositives + falsePositives)
- recall: truePositives / Positives
- f1-score: 2 * (precision * recall) / (precision + recall)
- support: Cantidad de ocurrencias

```python
print(metrics.classification_report(prediction, real_labels))
```

### Error samples:

```python
misslabels = prediction[real_labels != prediction]
reallabels = real_labels[real_labels != prediction]
missimages = mnist.test.images[real_labels != prediction]
choice = np.random.choice(range(len(misslabels)), 12, replace=False)
plt.figure(figsize=(16, 6))
for index, i in enumerate(choice):
    plt.subplot(2, 6, index+1)
    plt.axis('off')
    plt.imshow(missimages[i].reshape(28, 28), cmap=plt.cm.gray_r, interpolation='nearest')
    plt.title('Prediction: %i\nTrue: %i' % (misslabels[i], reallabels[i]))
```

Save results:

```python
df = pd.DataFrame(metrics.confusion_matrix(prediction, real_labels, range(10)))
df.to_csv('mlp.csv')
with open('mlpTime.txt', 'w') as file_:
    file_.write('%.4f' % elapsed)
```
